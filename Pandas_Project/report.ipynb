# Data that we transformed
In order to have concise tables that directly addressed our research questions, we took subsets from our organized versions of Gender_StatsCountry.csv and Gender_StatsData.csv. 
For our first question, we specificaly looked at data dealing with education completion and progression across countries. We chose to include the columns holding data on government spending on education, expected years of school, primary school completion rates for male and female, lower-secondary completion rates for male and female, and upper-secondary completion rates for male and female from Gender_StatsData.csv. In order to analyse our data by region we merged our subset from Gender_StatsData.csv with a subset from Gender_StatsCountry.csv containing only the country code and region for each country. From there, made a column that represented the sex for each row. Our final data set had independant variables: 'Country Code', 'Country Name', 'Year',and 'Sex' mapping to dependant variables: 'Gov. spending on education(% of GDP)', 'Expected Years of School', 'Primary School Completion Rate','Lower Secondary School Completion Rate', 'Upper Secondary School Completion Rate','Post Secondary School Completion Rate', 'Region'. 

For our second question,  we specificaly looked at unemployment and business related statistics. We chose to take of subset from Gender_StatsData.csv with information on the male and female unemployment rate, the share of female business owners and directors, number of female and male steps to register a business, and if women can work in the same industries as men. We took steps similar to question 1 in merging the unemployment data with 'Country Code' and 'Region'. We then split this table into two, one dealt with data that was measured for both sexes and the other table held data on females specifically. The first table had independant variables: 'Country Code','Country Name','Year',and 'Sex' mapping to dependant variables: 'Unemployment Rate','Steps to Start a Business','Region'. Our second table had independant variables:'Country Code','Country Name',and 'Year'  mapping to dependant variables: 'Share of Female Business Owners','Share of Female Directors','Women Work in Same Industry as Men (bool)','Region'.

For our thrid question, we specifacally looked at whether countries had specific legislation addressing sexual assualt and the prevelance of sexual assault. We took a subset of the Gender_StatsData.csv that contained one column on the % of women who had experienced sexual violence in the last 12 month and one boolean column that identified if a country had specific legislation addressing sexual violence. Continuing with the theme at looking at data with respect to geographical regions, we merged 'Region' into our dataset. The final table had independant variables 'Country Code','Country Name', and 'Year' mapping to dependant values: 'Legislation Addresses Sexual Violence','Prevelence of Sexual Violence', and 'Region'.
